Ques1. Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).
Ans:Key Differences and Relationships:
1.Artificial Intelligence (AI): The overarching field aiming to create
smart machines that can automate decision-making, simulate human behavior, and improve efficiency. It is the goal or the "big idea".
2.Machine Learning (ML): A method or subset of AI that uses statistical models to teach computers to learn from data rather than explicit programming. Examples include fraud detection or recommendation systems.
3.Deep Learning (DL): A specialized subset of Machine Learning based on artificial neural networks with multiple layers. It is used to handle vast, complex datasets, such as in image recognition, speech recognition, and natural language processing.
4.Data Science (DS): A multidisciplinary field focused on extracting actionable insights, patterns, and knowledge from structured or unstructured data. It involves data cleaning, visualization, and interpretation, often using AI/ML techniques, but often requires a "human in the loop" to interpret results.

Relationship Analogy: Data Science can be seen as the overall process, which uses AI/ML tools to find insights, while AI is the broader "intelligence" goal, ML is the method of learning, and DL is the advanced technique within that learning.

Ques2.What are the types of machine learning? Describe each with one
real-world example?
Ans:1. Supervised Learning (Predictive Learning)
Definition: Algorithms are trained on labeled data, meaning the input data is paired with the correct output, allowing the model to predict outcomes for new data.
Real-World Example: Email Spam Filtering. A service like Gmail learns from thousands of emails labeled "spam" or "not spam," enabling it to correctly classify new incoming emails.
2. Unsupervised Learning (Pattern Discovery)
Definition: Algorithms analyze unlabeled data, finding hidden structures, patterns, or groupings within data without explicit instructions on what to look for.
Real-World Example: Customer Segmentation. E-commerce sites analyze customer purchase history, behavior, and demographics to cluster shoppers into groups for targeted marketing, even if they don't know the exact categories beforehand.
3.Reinforcement Learning (Reward-Based)
Definition: An agent learns to make decisions by performing actions in an environment, receiving rewards for good actions and penalties for bad ones, iteratively optimizing behavior.
Real-World Example: Autonomous Vehicles. A self-driving car learns to navigate, obey traffic rules, and avoid obstacles, receiving positive reinforcement for safe driving and negative feedback for risky behavior.

Ques3. Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.
Ans:Key Definitions
1.Overfitting (High Variance, Low Bias): The model is too complex and fits training data perfectly but fails to generalize. Synonyms include high variance or memorization.
2.Underfitting (High Bias, Low Variance): The model is too simple and cannot capture the underlying patterns. Synonyms include high bias or oversimplification.
3.Bias-Variance Tradeoff: The fundamental struggle to balance model complexity. Reducing bias (underfitting) often increases variance (overfitting), and vice-versa. The goal is to reach the "sweet spot" where total error is minimized.

Ques4.What are outliers in a dataset, and list three common techniques for handling them?
Ans:Outliers are data points that significantly deviate from the rest of the observations in a dataset, representing anomalies, errors, or extreme variations. Synonyms include anomalies, anomalies, extremes, noise, or deviations.
Three Common Techniques for Handling Outliers:
1.Removal (Trimming): Deleting the outlier data points, which is suitable if they are errors or irrelevant to the analysis, though it risks losing important information.
2.Imputation: Replacing extreme outlier values with more representative values, such as the median, mean, or mode, to reduce their impact.
3.Winsorizing (Capping): Transforming the data by capping it at a certain percentile (e.g., 95th percentile), replacing outliers with the nearest "non-outlier" values.

Ques5.Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.
Ans.The Process of Handling Missing Values
1.Detection: Identifying missing entries, often using libraries like pandas (.isnull().sum()).
2.Analysis: Determining if data is Missing Completely at Random (MCAR), Missing at Random (MAR), or Missing Not at Random (MNAR) to decide the strategy.
3.Treatment/Strategy:
.Deletion: Dropping rows or columns with missing values (if minimal).
.Imputation: Replacing missing values with estimated values (mean, median, model-based).
.Flagging: Creating a new variable indicating the data was missing
Imputation Techniques
1.Numerical Data - Median Imputation: Replaces missing numerical values with the median of the non-missing values in that column. This is preferred over mean imputation when outliers are present.
2.Categorical Data - Mode/Most Frequent Imputation: Replaces missing categorical entries with the most frequently occurring category (mode) in that column.

Ques6.Write a Python program that:
● Creates a synthetic imbalanced dataset with make_classification() from
sklearn.datasets.
● Prints the class distribution.

```python
from sklearn.datasets import make_classification
from collections import Counter

# Create synthetic imbalanced dataset
X, y = make_classification(
    n_samples=1000,        # total samples
    n_features=10,         # number of features
    n_classes=2,           # binary classification
    weights=[0.9, 0.1],    # imbalance (90% vs 10%)
    random_state=42
)

# Print class distribution
print("Class distribution:")
print(Counter(y))
```

```python
"""Question 9: Implement Min-Max scaling on the following list of numbers [2, 5, 10, 15,
20] using sklearn.preprocessing.MinMaxScaler. Print the scaled array.
(Include your Python code and output in the code box below.)"""
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Given data
data = np.array([2, 5, 10, 15, 20]).reshape(-1, 1)

# Initialize scaler
scaler = MinMaxScaler()

# Fit and transform
scaled_data = scaler.fit_transform(data)

# Print result
print("Scaled Data:")
print(scaled_data)
```
